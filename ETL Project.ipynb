```python
import pandas as pd
import numpy as np
from itertools import chain

def chainer(s):
    return list(chain.from_iterable(s.str.split(',')))
```

```python
csv_file = "Data/MoviesOnStreamingPlatforms_updated.csv"
streaming_data=pd.read_csv(csv_file)
csv2_file = "Data/TMDb_updated.csv"
overview_data=pd.read_csv(csv2_file)
```

```python
streaming_data.head()
```

```python
streaming_data=streaming_data.drop(columns=['Unnamed: 0', 'Type','Language','Country'])
```

```python
streaming_data["Netflix"].replace({1: True, 0: False}, inplace=True)
streaming_data["Hulu"].replace({1: True, 0: False}, inplace=True)
streaming_data["Prime Video"].replace({1: True, 0: False}, inplace=True)
streaming_data["Disney+"].replace({1: True, 0: False}, inplace=True)
```

```python
streaming_data['Rotten Tomatoes']=streaming_data['Rotten Tomatoes'].str.replace(r'%', r'.0').astype('float')
streaming_data["Title"] = streaming_data["Title"].str.lower()
streaming_data["Title"] = streaming_data["Title"].str.replace(',','',regex=True).str.replace(':','',regex=True).str.replace('&','',regex=True).str.replace('and','',regex=True).str.replace('the','',regex=True).str.replace('  ',' ',regex=True)
```

```python
streaming_data=streaming_data.fillna('Null')
```

```python
streaming_data
```

```python
overview_data.head()
```

```python
overview_data=overview_data.drop(columns=['Unnamed: 0', 'original_language','vote_count','vote_average','Unnamed: 6'])
```

```python
overview_data["title"] = overview_data["title"].str.lower()
overview_data["title"] = overview_data["title"].str.replace(',','',regex=True).str.replace(':','',regex=True).str.replace('&','',regex=True).str.replace('and','',regex=True).str.replace('the','',regex=True).str.replace('  ',' ',regex=True)
```

```python
overview_data= overview_data.drop_duplicates(subset='title', keep="first")
```

```python
overview_data
```

```python
merge = streaming_data.merge(overview_data, left_on='Title',right_on="title")
```

```python
merge=merge[["ID","Title","Year","IMDb","Rotten Tomatoes","Netflix","Hulu","Prime Video","Disney+","Directors","Genres","Runtime","overview"]]
```

```python
merge.columns = map(str.lower, merge.columns)
```

```python
merge.rename(columns={"id": "movie_id","rotten tomatoes":"rotten_tomatoes","prime video":"prime_video","disney+":"disney_plus"},inplace=True)
```

```python
merge
```

```python
rating_df=merge[['movie_id','rotten_tomatoes','imdb']]
lookup=merge[['movie_id','title']]
streaming_service = merge[['movie_id','netflix','hulu','prime_video','disney_plus']]
overview_df=merge[['movie_id','overview']]
movie_details_df=merge[['movie_id','year','runtime']]
directors_df=merge[['movie_id','directors']]
genres_df = merge[['movie_id','genres' ]]
```

```python
lens = directors_df['directors'].str.split(',').map(len)

directors_normalized= pd.DataFrame({'movie_id': np.repeat(directors_df['movie_id'], lens),
                                   'directors': chainer(directors_df['directors'])})

directors_normalized
```

```python
director_names=directors_normalized[['directors']]
director_names.drop_duplicates(inplace=True)
director_names
director_names['director_id'] = range(1, len(director_names) + 1)

director_names.rename(columns={"directors": "director"})
```

```python
director_merge = directors_normalized.merge(director_names, left_on='directors',right_on="directors")
director_lookup=director_merge[['movie_id','director_id']]
director_lookup
```

```python
lens2 = genres_df['genres'].str.split(',').map(len)

genres_normalized= pd.DataFrame({'movie_id': np.repeat(genres_df['movie_id'], lens2),
                                   'genres': chainer(genres_df['genres'])})

genres_normalized
```

```python
genre_type=genres_normalized[['genres']]
genre_type.drop_duplicates(inplace=True)
genre_type['genre_id'] = range(1, len(genre_type) + 1)

genre_type.rename(columns={"genres": "genre_type"})
```

```python
genre_merge = genres_normalized.merge(genre_type, left_on='genres',right_on="genres")
genre_lookup=genre_merge[['movie_id','genre_id']]
genre_lookup
```
